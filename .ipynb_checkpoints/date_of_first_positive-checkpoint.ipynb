```python
import pandas as pd
from datetime import date
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('Data/clean_data.csv')
file
```

```python
dates = file[['month','death_yn','age_group']]

deaths = dates.groupby(dates['month']).count()

ages = dates.groupby([dates['month'],dates['age_group']]).sum()
age_story = dates.groupby([dates['month'],dates['age_group']]).count()
age_story = age_story.unstack()
ages = ages.unstack()
ages
```

```python
plt.bar(deaths.index,deaths['age_group'])
```

```python
ax = ages.plot.bar(stacked=True,figsize=(15,8))
plt.legend(loc='upper left')
```

```python
ax = age_story.plot.bar(stacked=True,figsize=(15,8))
plt.legend(loc='upper left')
```
